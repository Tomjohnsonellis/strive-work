#### 16. How to add a border (filled with 0's) around an existing array? (★☆☆)

````python
enclosed_array = np.ones([4,6])
print("Start:")
print(enclosed_array)
print()
print("=====")
print()

# To generate some 0s we need to know how many we need
# This will store the dimensions of the array to enclose
shape = np.shape(enclosed_array)

# This adds a column of zeros onto the end and start of our original array 
enclosed_array = np.column_stack((enclosed_array, np.zeros([shape[0]])))
enclosed_array = np.column_stack((np.zeros([shape[0]]) , enclosed_array))

# The shape has changed because we added 2 columns
shape = np.shape(enclosed_array)
# We just need to add a row of zeros at the start and end of the original array
top_and_bottom = np.zeros(shape[1])

# I used two different methods here just for fun, vstack just "stacks" rows together,
# Whereas insert allows you to choose exactly where to input the data
enclosed_array = np.vstack((enclosed_array, top_and_bottom))
enclosed_array = np.insert(enclosed_array, 0, 0, axis=0)

print("Finally:")
print(enclosed_array)
````

#### 17. What is the result of the following expression? (★☆☆)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
# So first, we are multiplying 0 by "nan" which is not a number, so this will result in "nan"
0 * np.nan
print(0 * np.nan) # nan

# Next we are seeing if nan is equal to itself, which I thought would be true BUT
# a nan is just a lack of data and by definition, nan != nan
np.nan == np.nan
print(np.nan == np.nan) # False

# This is attempting to compare positive infinity to not-a-number, resulting in false
np.inf > np.nan
print(np.inf > np.nan) # False

# Can we take away one nan from another? They aren't numbers so it just results in nan
np.nan - np.nan
print(np.nan - np.nan) # nan

# This one we can do, as we are seeing if a nan exists in a set containing a nan (which it does)
np.nan in set([np.nan])
print(np.nan in set([np.nan])) # True

# Finally some arithmetic, this is True in real life,
# HOWEVER: Due to how decimal numbers are handled by computers, comparing them is  generally unreliable.
0.3 == 3 * 0.1
print(0.3 == 3 * 0.1) # False


# These statements are independant of each other so the final result of these expressions is
# False
print("=====")
print(False)
````
