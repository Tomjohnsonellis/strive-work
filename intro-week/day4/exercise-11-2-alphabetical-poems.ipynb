Exercise 11.1

Write a program that reads the contents of the file "blakepoems.txt", splits it into words (where everything that is not a letter is considered a word boundary), and case-insensitively builds a dictionary that stores for every word how often it occurs in the text. Then print all the words with their quantities in alphabetical order.



## What do we need to do?

Split the words in a file and count how many times each one appears
(CaSe doEsn't maTTer)
Then print an ALPHABETICALLY sorted list of this information.

### How do we do it?

1. Read in the file
2. Split out the text
3. Format the text so that it is all the same case (lower)
4. Go through each word and either add it to a dictionary or, if it already exists, increment its value by 1
5. Print out how many times each word appears

```python
# poems = open("blakepoems.txt","r")
# print(poems.read())
# poems.close

# Originally I tried to look for any occurences of these punctuation symbols, then remove
# But "isalnum()" is a neater way of solving this issue
#punctuation = ('''!()-[]{};:'"\,<>./?@#$%^&*_~''')


# Firstly, open the file and call it 'poems'
with open("blakepoems.txt","r") as poems:
    
    # Next, read in the data and make it lowercase
    data = poems.read()
    data = data.lower()
    
    # Split up all the words into a list
    split_words = data.split()
    
    # Initialise a list for the clean words
    clean_words = []
    
    # For every word in our list of words
    for word in split_words:
        #print(f" Using: {word}")
        
        
        # For every character in that word
        for character in word:
            #print(f"--Using: {character}")
            
            # If it's something that isn't letters or numbers, remove it
            if character.isalnum() == False:
                #print(f"##### PUNCTUATION FOUND: {character}")
                word = word.replace(character, "")
                #print(f"===== Corrected word: {word}")
                
        # Add the fixed word to the list of clean words
        clean_words.append(word)

# # To check each word is actually cleaned up
# for element in clean_words:
#     print(element)


# Initialise the dictionary we need
counting_dictionary = {}  


# Go through each of our words and add it to the dictionary
# OR add 1 to the count if we've already added it
for word in clean_words:
    
    if word in counting_dictionary:
        counting_dictionary[word] += 1
    else:
        counting_dictionary[word] = 1


# Make a sorted version of the dictionary keys
sorted_dictionary = sorted(counting_dictionary)


# Finally, print out some results
print("Word --- Number of occurences")
print("")
for item in sorted_dictionary:
    print(f"{item} --- {counting_dictionary[item]}")
    

```

# Exercise 11.2

Do the same thing as you did for the previous exercise, but now process the text line by line. This is something that you would have to do if you had to process a very long text (as it is the case).

```python
# poems = open("blakepoems.txt","r")
# print(poems.read())
# poems.close

# Initialise the empty things we need
clean_words = []
counting_dictionary = {}


# Open the file and call it 'poems'
with open("blakepoems.txt","r") as poems:
    
    # Process a line until we run out
    while True:
        current_line = poems.readline()
        if not current_line:
            break
        
        # Make the text lowercase
        current_line = current_line.lower()
        # Split it up
        split_words = current_line.split()
        
        
        # Go through each character in each word, then remove anything we don't want
        for word in split_words:
            for character in word:
                if character.isalnum() == False:
                    word = word.replace(character, "")
            
            # Finally add the cleaned word to a list
            clean_words.append(word)
            # We've processed 1 line
            

# Fill in our empty dictionary
for word in clean_words:
    
    if word in counting_dictionary:
        counting_dictionary[word] += 1
    else:
        counting_dictionary[word] = 1


# Make a sorted version of the dictionary keys
sorted_dictionary = sorted(counting_dictionary)


# Finally, print out some results
print("Word --- Number of occurences")
print("")
for item in sorted_dictionary:
    print(f"{item} --- {counting_dictionary[item]}")     
    
```
